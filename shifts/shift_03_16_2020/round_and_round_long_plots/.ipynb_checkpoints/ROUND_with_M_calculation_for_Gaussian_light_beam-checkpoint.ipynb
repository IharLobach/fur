# Calculation of M

```python
lmda = 1.097
k0 = 2*np.pi/lmda
Nu=10.5
usw = k0/Nu
v = und_summary.loc['Middle', ['Beta_cm_X', 'Beta_cm_Y',
                              'Alpha_X', 'Alpha_Y',
                              'Angle_spread_rad_X','Angle_spread_rad_Y',
                              'ex_um', 'ey_um',
                              'Dispersion_cm_X', 'dDx/dS',
                              'dp/p',
                              'Sigma_um_X', 'Sigma_um_Y']]
bx, by, ax, ay, sxp, syp, ex, ey, Dx, Dxp, sp, sx, sy = v
bx, by, Dx = 1e4*np.array([bx, by, Dx])
```

```python
gx, gy = (1+ax**2)/bx, (1+ay**2)/by
Sx = np.sqrt(ex/gx+(gx*Dx+Dxp*ax)**2*bx*ex*sp**2/sxp**2)
Sy = np.sqrt(ey/gy)
dx = (ax*ex-Dx*Dxp*sp**2)/sxp**2
dy = ay/gy
```

## Gaussian light beam and nonzero electorn beam divergence

```python
M_Gauss_with_beam_div=\
    np.sqrt(1+4*usw**2*st**2)\
    *np.sqrt(1+4*k0**2*Sx**2*usxp**2+sxp**2*(4*k0**2*(usxp**2*dx**2+Sx**2)+1/usxp**2))\
    *np.sqrt(1+4*k0**2*Sy**2*usyp**2+syp**2*(4*k0**2*(usyp**2*dy**2+Sy**2)+1/usyp**2))
M_Gauss_with_beam_div
```

## Gaussian light beam and zero electron beam divergence

```python
M_Gauss_no_beam_div=\
    np.sqrt(1+4*usw**2*st**2)\
    *np.sqrt(1+4*k0**2*sx**2*usxp**2)\
    *np.sqrt(1+4*k0**2*sy**2*usyp**2)
M_Gauss_no_beam_div
```
