```python
direct_path = waveforms_dir.fi("RefCurve_2020-03-16_1_130532.Wfm.bin")
comb_path = waveforms_dir.fi("RefCurve_2020-03-16_10_132747.Wfm.bin")
```

```python
wf_dir= read_waveform(direct_path, one_channel=True)
```

```python
ch1, ch2 = read_waveform(comb_path)
ch1 = ch1[1:]
ch2 = ch2[1:]
```

```python
period_dir = get_period(wf_dir)
```

```python
period_ch2 = get_period(ch2)
```

```python
period_dir, period_ch2
```

```python
n = -1
res = []
for waveform, per in zip([wf_dir, ch2, ch1],[period_dir, period_ch2, period_ch2]):
    wf = waveform[:n]
    dir_df = pd.DataFrame({'t': np.arange(len(wf)), 'V': wf})
    dir_df['t_mod'] = dir_df['t'] % per
    dir_df['t_bin'] = dir_df['t_mod'].astype(int)
    res.append(dir_df.groupby('t_bin').mean()['V'].values)
direct, comb, comb_delta = res
```

```python
plt.plot(wf_dir[:3000])
plt.plot(direct)
```

```python
plt.plot(ch2[:3000])
plt.plot(comb)
```

```python
plt.plot(ch1[1:3000])
plt.plot(comb_delta)
#plt.plot(comb_delta-np.mean(comb_delta))
```

```python
direct0 = np.roll(direct,-610)
plt.plot(direct0)
plt.plot(0.65*(direct0-np.mean(direct0)))
plt.plot(comb)
```

```python
freqs = np.fft.rfftfreq(len(comb), d=40e-3)
transm_f = np.fft.rfft(comb)/np.fft.rfft(direct0)
fig, ax = plt.subplots(figsize=[15, 7.5])
ax.plot(
    freqs,
    np.absolute(transm_f))
ax.set_xlim(0, 0.1)
ax.set_ylim(0, 1)
ax.set_xlabel('GHz')
ax.set_ylabel('linear transformation function (abs)')
ax.set_title('transformation: Direct signal -> Comb Sum output')
```

```python
fig, ax = plt.subplots(figsize=[15, 7.5])
n = 5000
d = 20
ax.plot(100*ch1[1:n])
ax.plot(ch2[1:n])
diff0 = np.roll(direct0,-d)-direct0
ax.plot(diff0)
diff_transformed = np.fft.irfft(np.fft.rfft(diff0)*transm_f, n=len(diff0))
ax.plot(5*diff_transformed+0.5*comb)
```

# Using model signal and model low pass filter

```python
amps = np.random.normal(loc=1, scale=0.001, size=10000)
model_direct = np.ravel(amps[:, None]*direct0)
model_delayed = np.roll(model_direct,len(direct0))
model_comb_sum = model_direct+model_delayed
model_comb_diff = model_direct-model_delayed
```

```python
period = get_period(model_comb_sum)
output = {}
sum_amp, diff_var,*_ = get_fluctiation_and_noise_var(model_comb_diff, model_comb_sum, period, n_bins=1000,
                              show_plots=True, fit_method='lstsq',
                              output_dic=output)
np.sqrt(diff_var)/sum_amp
```

```python
f0 = 0.002 #GHz
model_freqs =  np.fft.rfftfreq(len(model_direct),d=40e-3)
model_transm = 1j*model_freqs/f0/(1+1j*model_freqs/f0)
```

```python
comb_sum_filtered = np.fft.irfft(np.fft.rfft(model_comb_sum)*model_transm)
comb_diff_filtered = np.fft.irfft(np.fft.rfft(model_comb_diff)*model_transm)
plt.plot(comb_sum_filtered[:20000])
plt.show()
plt.plot(comb_diff_filtered[:20000])
```

```python
period = get_period(comb_sum_filtered)
output = {}
sum_amp, diff_var,*_ = get_fluctiation_and_noise_var(comb_diff_filtered, comb_sum_filtered, period, n_bins=1000,
                              show_plots=True, fit_method='lstsq',
                              output_dic=output)
np.sqrt(diff_var)/sum_amp
```
