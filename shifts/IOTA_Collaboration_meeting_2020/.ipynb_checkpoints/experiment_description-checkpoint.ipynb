```python
numRev = int(1.5e-3/133e-9)
revNum = np.arange(numRev)
#numPh = np.random.poisson(lam=0.1, size=numRev)
numPh = np.zeros(numRev, dtype=np.int)
for i in range(numRev):
    if i == 0:
        numPh[i] = 1
    elif i == 3:
        numPh[i] = 1
    elif i == 4:
        numPh[i] = 2
df = pd.DataFrame({"Number of photocounts, N":numPh})
df.index.name = "Revolution number"
df
```
