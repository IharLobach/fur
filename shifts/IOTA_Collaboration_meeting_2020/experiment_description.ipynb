```python
numRev = int(1.5e-3/133e-9)
revNum = np.arange(numRev)
numPh = np.random.poisson(lam=int(1e7), size=numRev)
df = pd.DataFrame({"Revolution number":revNum, "Number of photocounts, N":numPh})
df
```
