```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f0 = 7.5e6
T0 = 1/f0
dt = 0.05e-9
t0 = np.arange(0, T0, dt)
def get_y(t):
    return np.exp(-((t % T0)-60e-9)**2/2/7e-9**2)
y0 = get_y(t0)
```

```python
fig, ax = plt.subplots()
ax.plot(t0, y0)
ax.set_title("One-turn gaussian signal")
ax.set_xlabel("Time (s)")
plt.show()
```

```python
n = 10000
t = np.arange(0, n*T0, dt)
y = get_y(t)
freq = np.fft.rfftfreq(len(y), dt)
amp = np.abs(np.fft.rfft(y))
```

```python
fig, ax = plt.subplots(figsize=(15, 7))
ax.plot(freq, amp)
ax.set_yscale('log')
ax.set_xscale('log')
ax.set_ylabel('FFT Amplitude')
ax.set_xlabel('FFT Frequancy (Hz)')
plt.show()
```

```python
freq0 = 7.5e6
dfreq = 0.1e6
indices = (freq > freq0 - dfreq) & (freq < freq0 + dfreq)
fig, ax = plt.subplots(figsize=(15, 7))
ax.plot(freq[indices], amp[indices])
ax.set_yscale('log')
ax.set_xscale('log')
ax.set_ylabel('FFT Amplitude')
ax.set_xlabel('FFT Frequancy (Hz)')
plt.show()
```
