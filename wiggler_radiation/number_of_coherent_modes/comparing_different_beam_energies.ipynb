```python
from wiggler_radiation.number_of_coherent_modes.coherent_modes import get_M_interpolator_sxsyEn
import matplotlib.pyplot as plt
import numpy as np
```

```python
M_interps = get_M_interpolator_sxsyEn()
```

```python
sx = 1200
sy = 100
sz = 30
energies = np.linspace(95, 105, 100)
Ms = [M_interps[0](sx,sy,sz,en) for en in energies]
fig, ax = plt.subplots()
ax.plot(energies, Ms)
ax.set_ylim(0,ax.get_ylim()[1])
ax.set_xlabel("Beam energy, MeV")
ax.set_ylabel("Nubmer of coherent modes M")
ax.set_title("Flat beam")
plt.show()
```

```python
Ms = [M_interps[-1](sx,sy,sz,en) for en in energies]
fig, ax = plt.subplots()
ax.plot(energies, Ms)
ax.set_xlabel("Beam energy, MeV")
ax.set_ylabel("dM/dE, 1/MeV")
ax.set_title("Flat beam")
plt.show()
```
